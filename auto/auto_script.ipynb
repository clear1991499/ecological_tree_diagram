````python
import csv
````

````python
read_file = "auto.csv"
write_file = "auto.md"
````

````python
def set_dict(d, key, data, rank, i, j):
    text = data[i][j]
    text += "" if (text in [" ", "不明"]) or (rank[j] == "種") else rank[j]
    d.setdefault(key, text)
````

````python
with open(read_file) as f:
    reader = csv.reader(f)
    l = [row for row in reader]
rank = l[0]
data = l[1:]

for i in range(len(data)):
    for j in range(len(data[i])):
        if data[i][j] == "-":
            data[i][j] = " "

d = {}

ids = [["a"*(i+1) for i in range(len(data[0]))]]
for i, k in enumerate(ids[0]):
    set_dict(d, k, data, rank, 0, i)

for i in range(1, len(data)):
    print_list = []
    id = []
    div = False
    for j in range(len(data[i])):
        if data[i][j] == "=":
            id += [ids[i-1][j]]
            data[i][j] = data[i-1][j]
        elif data[i][j] != "=":
            if not div:
                prev = ids[i-1][j]
                id += [prev[:-1] + chr(ord(prev[-1])+1)]
                div = True
            else:
                id += [id[-1]+"a"]        
            set_dict(d, id[-1], data, rank, i, j)

        text = "" if data[i][j] == " " else data[i][j]
        text += "" if (text in ["", "不明"]) or (rank[j] == "種") else rank[j]

        if text != "":
            print_list += [text]
    ids += [id]

    print(" ".join(print_list))
#for id in ids:
#    print(id)
#d
````

````python
with open(write_file, mode="w") as f:
    f.write("```mermaid\n")
    f.write("graph LR\n")
    f.write(" --- ".join(rank)+"\n")
    for i, id in enumerate(ids):
        if i == 0:
            f.write(" --- ".join(id)+"\n")
        else:
            start = 0
            for j in range(len(id)):
                if ids[i][j] != ids[i-1][j]:
                    start = max(start, j-1)
                    break
            f.write(" --- ".join(id[start:])+"\n")
    for k, v in d.items():
        f.write(f"{k}[\"{v}\"]\n")
    f.write("```\n")
````
